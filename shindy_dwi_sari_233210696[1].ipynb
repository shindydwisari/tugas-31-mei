```python
import seaborn as sns 
import matplotlib.pyplot as plt

# Data contoh
usia = [20,22,21,23,20,22,21,24,23,20]
ipk = [3.5,3.2,3.8,2.9,3.1,3.6,2.7,3.9,3.3,3.2]

# scatter plot #1
plt.figure(figsize=(6,4))
sns.scatterplot(x=usia, y=ipk)
plt.title('Hubungan Antara Usia dan Ipk #1')
plt.xlabel('Usia')
plt.ylabel('Ipk')
plt.grid(True)
plt.show()

# scatter plot #2
plt.figure(figsize=(6,4))
sns.regplot(x=usia, y=ipk)
plt.title('Hubungan Antara Usia dan Ipk #2')
plt.xlabel('Usia')
plt.ylabel('Ipk')
plt.grid(True)
plt.show()
```

```python
PENJELASAN:
Pada codingan ini mengimpor dua library, seaborn dan matplotlib.pyplot, yang digunakan untuk visualisasi data. 
Seaborn menyediakan gambar grafik statistik menarik, sedangkan Matplotlib adalah library dasar untuk plotting di Python.
Data berisikan Usia dan ipk
plt.figure(figsize=(6, 4)): Membuat sebuah figure baru dengan ukuran 6x4 inci.
sns.scatterplot(x=usia, y=ipk): Menggambar scatter plot menggunakan Seaborn, dengan usia sebagai sumbu x dan ipk sebagai sumbu y.
plt.title('Hubungan Antara Usia dan IPK #1'): Menambahkan judul pada plot.
plt.xlabel('Usia'): Memberi label sumbu x dengan teks "Usia".
plt.ylabel('IPK'): Memberi label sumbu y dengan teks "IPK".
plt.grid(True): Menampilkan grid pada plot.
plt.show(): Menampilkan plot.
Kode ini menghasilkan dua plot:
    Plot pertama adalah scatter plot sederhana yang menunjukkan hubungan antara usia dan Ipk.
    Plot kedua adalah scatter plot dengan garis regresi yang menunjukkan tren atau pola hubungan antara usia dan Ipk.
```

#CONTOH 5

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# Data contoh
usia = [20,22,21,23,20,22,21,24,23,20]
ipk = [3.5,3.2,3.8,2.9,3.1,3.6,2.7,3.9,3.3,3.2]

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(usia, bins=20, kde=True, color='red')
plt.title('Distribusi Usia')
plt.xlabel('usia')
plt.ylabel('ipk')
plt.grid(True)
plt.show()
```

```python
PENJELASAN: 
Pada codingan ini mengimpor dua library, seaborn dan matplotlib.pyplot, yang digunakan untuk visualisasi data. 
Seaborn menyediakan gambar grafik statistik menarik, sedangkan Matplotlib adalah library dasar untuk plotting di Python.
Data berisikan Usia dan ipk
plt.figure(figsize=(6, 4)): Membuat figure baru dengan ukuran 6x4 inci.
sns.histplot(usia, bins=20, kde=True, color='green'): Menggambar histogram menggunakan Seaborn.
    usia: Data yang akan diplot dalam histogram.
    bins=20: Menentukan jumlah bin (interval) dalam histogram.
    kde=True: Menambahkan estimasi densitas kernel (garis halus) pada histogram untuk memperlihatkan distribusi data.
    color='red': Mengatur warna histogram menjadi hijau.
    plt.title('Distribusi Usia'): Menambahkan judul pada plot dengan teks "Distribusi Usia".
    plt.xlabel('Usia'): Memberi label sumbu x dengan teks "Usia".
    plt.ylabel('Frekuensi'): Memberi label sumbu y dengan teks "Frekuensi". (Label ini seharusnya "Frekuensi" karena
    histogram menunjukkan frekuensi, bukan IPK).
    plt.grid(True): Menampilkan grid pada plot untuk mempermudah pembacaan data.
    Frekuensi Usia: Anda dapat melihat seberapa sering setiap rentang usia muncul dalam dataset Anda.
    Kepadatan Usia: Garis KDE (Kernel Density Estimate) memperlihatkan bentuk distribusi dari data usia Anda, 
codingan ini membantu mengidentifikasi apakah data mengikuti distribusi normal atau memiliki bentuk lain (misalnya, distribusi miring
atau multimodal).Dengan membuat histogram ini, Anda dapat lebih memahami karakteristik data usia Anda dan menggunakan informasi ini untuk 
analisis lebih lanjut atau pembuatan model yang lebih akurat.
```

#CONTOH 1

```python
import matplotlib.pyplot as plt

# Data
gender = [1, 0, 1, 1, 0, 1, 1, 0, 1, 1]
ipk = [3.5, 3.2, 3.8, 2.9, 3.1, 3.6, 2.7, 3.9, 3.3, 3.2]

# Plot
plt.figure(figsize=(10, 6))

# Graduation status 1
ipk_lulus = [ipk[i] for i in range(len(gender)) if gender[i] == 1]
# Graduation status 0
ipk_tidak_lulus = [ipk[i] for i in range(len(gender)) if gender[i] == 0]

plt.scatter([1]*len(ipk_lulus), ipk_lulus, color='green', label='Lulus (1)')
plt.scatter([0]*len(ipk_tidak_lulus), ipk_tidak_lulus, color='black', label='Tidak Lulus (0)')

# Labels and title
plt.xlabel('Status Kelulusan')
plt.ylabel('IPK')
plt.title('Hubungan antara gender dan IPK')
plt.legend()
plt.xticks([0, 1], ['Tidak Lulus', 'Lulus'])
plt.grid(True)

# Show plot
plt.show()
```

```python
PENJELASAN:
Pada codingan ini mengimpor library matplotlib.pyplot untuk membuat visualisasi data.
Data yang diguanakan gender dan ipk
Membuat figure baru dengan ukuran 10x6 inci.
Menggunakan list comprehension untuk memisahkan nilai IPK berdasarkan gender. 
ipk_lulus berisi IPK untuk yang lulus (gender=1), dan ipk_tidak_lulus berisi IPK untuk yang tidak lulus (gender=0).
Membuat scatter plot dengan sumbu x menunjukkan status kelulusan (1 untuk lulus, 0 untuk tidak lulus) dan sumbu y 
menunjukkan IPK. Poin plot diwarnai sesuai dengan status kelulusan.
Mengatur label sumbu, judul, legenda, dan tanda sumbu x menjadi 'Tidak Lulus' dan 'Lulus'. Menampilkan grid pada plot.
Codingan yang Anda tulis bertujuan untuk membuat scatter plot yang menunjukkan hubungan antara gender dan IPK.

```
